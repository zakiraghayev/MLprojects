```python
data = data.drop(['Orbiting Body','Neo Reference ID', 'Name', 'Equinox', 'Orbit Determination Date', 'Close Approach Date' ], axis=1)
data.head()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import RandomizedSearchCV
from skopt import BayesSearchCV
```

```python

# Create Dummy variables
data_prep = pd.get_dummies(data, drop_first=True)

# Create X and Y Variables
X = data_prep.iloc[:, :-1]
Y = data_prep.iloc[:, -1]
X
```

```python
xbm = XGBClassifier(verbosity = 0)

xbm_param = {
    "n_estimators":[9, 10, 11, 12 ,13],
    "max_depth":[4,5,6,7,8,9,10],

}
```

```python

# Create the RandomizedSearchCV object
xbm_rs = RandomizedSearchCV(estimator=xbm, 
                        param_distributions=xbm_param,
                        scoring='accuracy',
                        cv=10,
                        n_iter=10,
                        return_train_score=True,
                        random_state=1234)

# n_iter selects 10 combinations out of 30 possible
# Now 10 x 10 = 100 jobs will be executed

# Fit the data to RandomizedSearchCV object
xbm_rs_fit = xbm_rs.fit(X, Y)

# Get the results of RandomizedSearch
cv_results_xbm_rs = pd.DataFrame.from_dict(xbm_rs_fit.cv_results_)

```

```python
# Print the best parameters of Randomized Search for Random Forest
print('\n The best Parameters are : ')
print(xbm_rs_fit.best_params_)
```

```python
# parameter ranges are specified by one of below
from skopt.space import Real, Categorical, Integer

from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X, Y,
#                                                     train_size=0.25,
#                                                     random_state=0)

# log-uniform: understand as search over p = exp(x) by varying x
opt = BayesSearchCV(
    estimator=xbm,
    search_spaces=xbm_param,
    n_iter=32,
    random_state=0
)

# executes bayesian optimization
opt_fit = opt.fit(X, Y)

# model can be saved, used for predictions or scoring
# print(opt.score(X_test, y_test))
```

```python
# Print the best parameters of Randomized Search for BayesSearchCV
print('\n The best Parameters are : ')
print(opt_fit.best_params_)
```

```python
model = XGBClassifier(**{'n_estimators': 12, 'max_depth': 6})
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
