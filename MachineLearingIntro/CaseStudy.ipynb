```python
from datetime import datetime

class DatasTimeConverter:
    def __init__(self, df, formats=[ '%m/%d/%Y']  ):
        
        self.formats = formats
        self.df = df
        self.epoch = datetime.utcfromtimestamp(0)
        
    def unix_time_sec(self,dt):
        return (dt - self.epoch).total_seconds()
    
    def converter(self, date):
        return datetime.strptime(date, self.formats[0])
    # TODO: add 0 if time string is problem like in this case
    
    def convert(self):
        arr = [ self.unix_time_sec(self.converter(date) ) for date in self.df ]
        return  pd.DataFrame( np.array(arr) ).values.reshape(-1,1)
    
    
```

```python
# convert data to cumulative data
df_new = pd.DataFrame()
df_new = df[['Confirmed', 'Recovered', 'Deaths']].cumsum()
df_new.columns = ["Cum_Confirmed", "Cum_Recovered", "Cum_Death"]
df_new["ObservationDate"] = DatasTimeConverter( df["ObservationDate"] ).convert()
#df["ObservationDate"]

plt.plot(df_new['ObservationDate'], df_new["Cum_Confirmed"])
plt.plot(df_new['ObservationDate'], df_new["Cum_Death"])
plt.plot(df_new['ObservationDate'], df_new["Cum_Recovered"])
```

```python
model = LinearRegression()
```

```python
X = df_new[ "ObservationDate" ].values.reshape(-1,1)
y = df_new[ ["Cum_Death", "Cum_Confirmed", "Cum_Recovered"] ]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model.fit(X_train, y_train)
```

```python
arr = df_new['ObservationDate'].values

aday= arr[-1] - arr[-2]

new_perd = []
for i in range(1,8):
    day = arr[-1].reshape(-1,1)+aday*i
    new_perd.append( (day[0][0], 
                      round(model.predict(day)[0][0] ),  
                      round(model.predict(day)[0][1]), 
                      round(model.predict(day)[0][2])  ) )

pred_data = pd.DataFrame(new_perd)
print(new_perd)
pred_data.columns = ["ObservationDate","Cum_Confirmed", 'Cum_Death', 'Cum_Recovered' ]
# df_new.append(pred_data)

plt.plot(df_new['ObservationDate'], df_new["Cum_Confirmed"])
plt.plot(df_new['ObservationDate'], df_new["Cum_Death"])
plt.plot(df_new['ObservationDate'], df_new["Cum_Recovered"])
plt.plot(pred_data['ObservationDate'], pred_data["Cum_Confirmed"], color='red')
plt.plot(pred_data['ObservationDate'], pred_data["Cum_Death"], color='blue')
plt.plot(pred_data['ObservationDate'], pred_data["Cum_Recovered"], color='green')
```
